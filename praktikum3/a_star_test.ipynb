```python
from a_star import AStar
from tabulate import tabulate

class AStarForTests(AStar):
    def run(self) -> None:
        table = []
        for start in self.graph.keys():
            for ziel in self.graph.keys():
                self.start = start
                self.ziel = ziel
                res = self.astar()
                table.append([start, ziel, res.cost, " -> ".join(res.way)])
        print(tabulate(table, headers=["Von", "Nach", "Kosten", "Weg"], tablefmt="fancy_grid"))


AStarForTests().run()
```
